**Clean Order Dataset**

---




```python
order_df = pd.read_csv('olist_orders_dataset.csv')
order_df.isnull().sum()
```

```python
order_df.info()
```

```python
# Change Date data type

order_df['order_purchase_timestamp'] = pd.to_datetime(order_df['order_purchase_timestamp'])
order_df['order_approved_at'] = pd.to_datetime(order_df['order_approved_at'])
order_df['order_delivered_carrier_date'] = pd.to_datetime(order_df['order_delivered_carrier_date'])
order_df['order_delivered_customer_date'] = pd.to_datetime(order_df['order_delivered_customer_date'])
order_df['order_estimated_delivery_date'] = pd.to_datetime(order_df['order_estimated_delivery_date'])

order_df.info()

```

```python
order_df = order_df.dropna(subset=['order_delivered_customer_date'])
order_df.isnull().sum()
```

```python
# Add new column
# delivery days
order_df['delivery_days'] = (
    order_df['order_delivered_customer_date'] - order_df['order_purchase_timestamp']
).dt.days

# delivery status
order_df['delivery_status'] = order_df.apply(
    lambda x: 'Late' if x['order_delivered_customer_date'] > x['order_estimated_delivery_date'] else 'On Time',
    axis=1
)

order_df.head()
```

```python
order_df.isnull().sum()
```

```python
# Better Readability - change columns name

order_df = order_df.rename(columns={
    'order_purchase_timestamp': 'order_placed',
    'order_delivered_customer_date': 'delivered_date',
    'order_estimated_delivery_date': 'expected_delivery_date',
    'order_approved_at': 'approved_time'
})

order_df.head()
```

```python
# Make names: lowercase, no spaces, consistent
order_df.columns = order_df.columns.str.lower().str.strip()
order_df.head()
```

```python
# Change column values in title
order_df['order_status'] = order_df['order_status'].str.title()
```

```python
order_df['delivery_days'].describe()
```

```python
order_df['delivered_date'].isnull().sum()
```

```python
# Download the order dataset file
order_df.to_csv('order_df.csv', index=False)
files.download('order_df.csv')
```

```python
# Fix data type
reviews_df['review_creation_date'] = pd.to_datetime(reviews_df['review_creation_date'])
reviews_df['review_answer_timestamp'] = pd.to_datetime(reviews_df['review_answer_timestamp'])
reviews_df.info()
```

```python
# Create new column

def review_category(score):
  if score >= 4:
    return 'Positive'
  elif score == 3:
    return 'Neutral'
  else:
    return 'Negative'

reviews_df['review_category'] = reviews_df['review_score'].apply(review_category)
reviews_df.head()
```

**Product Dataset**
---

```python
from google.colab import files
uploaded = files.upload()
```

```python
products_df = pd.read_csv('products_dataset.csv')

# Check nulls
products_df.isnull().sum()
```

```python
# Remove the nulls values

products_df = products_df.dropna(subset=['product_category_name','product_weight_g',
                                         'product_length_cm', 'product_height_cm', 'product_width_cm'])
products_df.isnull().sum()
```

```python
# Check duplicates

products_df.duplicated().sum()
```

```python
# Drop unwanted columns

products_df = products_df.drop(columns=['product_name_lenght', 'product_description_lenght', 'product_photos_qty'])
products_df.info()
```

```python
# Change product category name (in Portuguese to English)

# First load the translate table
product_category_name_translation = pd.read_csv('product_category_name_translation.csv')

products_df = products_df.merge(product_category_name_translation, on = 'product_category_name', how= 'left')

products_df.isnull().sum()
```

```python
# After merge drop the Portuguses column

products_df = products_df.drop(columns=['product_category_name'])
products_df.head()
```

```python
# Rename the product category column

products_df = products_df.rename(columns={'product_category_name_english': 'product_category'})
products_df.head()
```

```python
# Change the values format

products_df['product_category'] = products_df['product_category'].str.title()
products_df.head()
```

```python
# Replace the underscore in category value

products_df['product_category'] = products_df['product_category'].str.replace('_', " ")
products_df.head()
```

```python
products_df['product_category'].value_counts()
```

```python
# Rearrange the columns order

products_df = products_df[
     [
        'product_id',
        'product_category',
        'product_weight_g',
        'product_length_cm',
        'product_height_cm',
        'product_width_cm'
    ]
]

products_df.head()
```

```python
products_df.info()
```

```python
# Download the order dataset file
products_df.to_csv('products_df.csv', index=False)
files.download('products_df.csv')
```
